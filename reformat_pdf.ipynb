```python
import re
from pdfminer.high_level import extract_text

def extract_references_from_pdf(pdf_path):
    # Extract text from the PDF
    text = extract_text(pdf_path)

    # Find the references section (this can be adjusted based on common patterns)
    references_start = re.search(r'\b(References|Bibliography|Works Cited)\b', text)
    
    if not references_start:
        print("References section not found.")
        return

    # Extract the references section (assuming the rest of the document after 'References' is the references section)
    references_text = text[references_start.start():]

    # Split references into individual entries (this is a basic split on numbers. You may need to adjust this regex)
    references = re.split(r'\n(?=\[?\d+\]?\.?)', references_text)

    return references[1:]  # the first item is the "References" title

pdf_path = "attention_paper.pdf"
references = extract_references_from_pdf(pdf_path)
for ref in references:
    print(ref)
    print('-' * 80)  # Separator
```

```python
#get item that looks like ****.****
def get_arxiv_number(reference):
    # Extract the arXiv number
    arxiv_number = re.search(r'arXiv:\d{4}\.\d{4,5}(v\d+)?', reference)
    # get rid of arxiv: in the string
    if arxiv_number:
        arxiv_number = arxiv_number.group(0)[6:]
        return arxiv_number
    ref_number = re.search(r'abs/\d{4}\.\d{4,5}(v\d+)?', reference)
    if ref_number:
        ref_number = ref_number.group(0)[4:]
        return ref_number
    return None
```

```python
def get_source_number(reference):
    number = re.search('\d{4}\.\d{4,5}(v\d+)?', reference)
    return number.group(0)

print(get_arxiv_number(references[8]))
```

```python
# clean text so that it is just words and spaces
# Remove words covered with newlines
# remove items after references
def clean_article(text):
    # removed words with preceding newlines and trailing newlines
    pattern = r'(^|\n)\S+(\n|$)'
    text = re.sub(pattern, '\n', text)
    text = text.strip()

    return text
```

```python
def keep_paragraphs(s):
    # Split the input into chunks by double newlines
    chunks = s.split('\n\n')
    
    # Keep only chunks that have more than one line or that have a certain length
    paragraphs = [chunk for chunk in chunks if '\n' in chunk or len(chunk.split()) > 5]
    
    # Join the paragraphs with double newlines and return
    return '\n\n'.join(paragraphs)

# Test
s = '''
This is a standalone line.

This is the start of a paragraph.
It continues here.

Another standalone line.

This is another paragraph.
'''
print(keep_paragraphs(s))
```

```python
def clean_paragraphs(s):
    # Split the input into chunks by double newlines
    chunks = s.split('\n\n')
    
    cleaned_paragraphs = []
    
    for chunk in chunks:
        # For each chunk, filter out lines that are too short or standalone characters
        cleaned_lines = [line for line in chunk.split('\n') if len(line.strip()) > 2]
        if cleaned_lines:
            cleaned_paragraphs.append('\n'.join(cleaned_lines))
    
    # Join the cleaned paragraphs with double newlines and return
    return '\n\n'.join(cleaned_paragraphs)

```
